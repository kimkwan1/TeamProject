```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.round(np.arange(0, 5.00001, 0.01), 2)
tmpT = []
nT = np.zeros_like(T)
mag = np.zeros_like(T)
sus = np.zeros_like(T)
bin_cum = np.zeros_like(T)

with open("./orders.csv", "r") as fid:
    line = fid.readline()
    print(line)
    line = fid.readline()
    while line:
        vals = line.split(",")
        tmpT.append(vals[0])
        # read id
        n = int(np.round(float(vals[0])*100))
        nT[n] += 1
        m = float(vals[1])
        s = float(vals[2])
        b = float(vals[3])

        mag[n] += m
        sus[n] += s
        bin_cum[n] += b

        line = fid.readline()

mag = mag / nT
sus = sus / nT
bin_cum = bin_cum / nT
```

```python
plt.figure(dpi=150)
plt.plot(T, mag, 'k.')
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(T, sus, 'r.')
plt.show()
```
