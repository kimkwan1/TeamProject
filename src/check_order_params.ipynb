```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.round(np.arange(0, 5.00001, 0.01), 2)
tmpT = []
nT = np.zeros_like(T)
mag = np.zeros_like(T)
sus = np.zeros_like(T)
bin_cum = np.zeros_like(T)

with open("./data/orders.csv", "r") as fid:
    line = fid.readline()
    print(line)
    line = fid.readline()
    while line:
        vals = line.split(",")
        tmpT.append(vals[0])
        # read id
        n = int(np.round(float(vals[0])*100))
        nT[n] += 1
        m = float(vals[1])
        s = float(vals[2])
        b = float(vals[3])

        mag[n] += m
        sus[n] += s
        bin_cum[n] += b

        line = fid.readline()

mag = mag / nT
sus = sus / nT
bin_cum = bin_cum / nT
```

```python
Tc = 2.269
```

## magnetization

```python
plt.figure(dpi=200, figsize=(8, 3))

plt.subplot(121)
plt.plot(T, mag, 'ko', fillstyle='none', markersize=3)
plt.vlines(Tc, -1, 2, color='r', linestyle='--')
plt.xlabel(r"k$_{b}$T/J", fontsize=12)
plt.ylabel("magnetization", fontsize=12)
plt.ylim([-0.05, 1.05])

plt.subplot(122)
plt.plot(T, sus, 'ko', fillstyle='none', markersize=3)
plt.vlines(Tc, -10, 120, color='r', linestyle='--')
plt.xlabel(r'k$_{b}$T/J', fontsize=12)
plt.ylabel('magnetic susceptibility', fontsize=12)
plt.ylim([-5, 105])

plt.tight_layout()
plt.show()
```
